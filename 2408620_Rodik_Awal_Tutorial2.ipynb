```python
import matplotlib.pyplot as plt

# Data for unequal intervals
weights = [5, 6, 8, 4, 2]  # Frequencies
intervals = [(0, 40), (40, 50), (50, 60), (60, 70), (70, 100)]  # Class intervals

# Calculate densities and bar midpoints
densities = [weights[i] / (intervals[i][1] - intervals[i][0]) for i in range(len(weights))]
widths = [interval[1] - interval[0] for interval in intervals]
midpoints = [(interval[0] + interval[1]) / 2 for interval in intervals]

# Plot histogram as a bar chart
plt.bar(midpoints, densities, width=widths, color='lightblue', edgecolor='black', align='center')
plt.title("Density-Adjusted Histogram")
plt.xlabel("Weight Intervals (grams)")
plt.ylabel("Density")
plt.xticks(midpoints, [f"{interval[0]}-{interval[1]}" for interval in intervals])  # Label x-axis with intervals
plt.show()
```
