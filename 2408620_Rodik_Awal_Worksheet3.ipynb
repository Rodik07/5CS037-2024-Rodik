**<font color = "Red"> Which passenger had the highest fare paid relative to their age?**

```python
# 1. Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of the passenger(i.e., Fare/Age).
# 2. Subset rows where fare per year is higher than 5, assigning this to high fare age.
# 3. Sort high fare age by descending fare per year, assigning this to high fare age srt.
# 4. Select only the Name and fare per year columns of high fare age srt and save the result as result.
# 5. Look at the result.

# Assuming titanic_data DataFrame is already loaded as in the previous code.
# Adding Titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Worksheet-2/Dataset week3/Copy of Titanic-Dataset.csv')

# 1. Adding fare per year containing the fare divided by the age of the passenger.
titanic_data['Fare_per_year'] = titanic_data['Fare'] / titanic_data['Age']

# 2. Subset rows where fare per year is higher than 5.
high_fare_age = titanic_data[titanic_data['Fare_per_year'] > 5]

# 3. Sort high fare age by descending fare per year.
high_fare_age_srt = high_fare_age.sort_values(by='Fare_per_year', ascending=False)

# 4. Select only the Name and fare per year columns.
result = high_fare_age_srt[['Name', 'Fare_per_year']].head(8)

# 5. Result
result
```

**<font color = "red">Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
class?**

```python
# 1. Add a column to the Titanic dataset, fare per class, containing the fare divided by the passenger class i.e. Fare / Pclass.
# 2. Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equal to 18), assigning this to adult males.
# 3. Sort adult males by descending fare per class, assigning this to adult males srt.
# 4. Select only the Name, Age, and fare per class columns of adult males sr and save the result
# as result.
# 5. Look at the result.

# Assuming titanic_data DataFrame is already loaded as in the previous code.
# Adding Titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Worksheet-2/Dataset week3/Copy of Titanic-Dataset.csv')

# 1. Add a column 'Fare_per_class'
titanic_data['Fare_per_class'] = titanic_data['Fare'] / titanic_data['Pclass']

# 2. Subset adult males
adult_males = titanic_data[(titanic_data['Sex'] == 'male') & (titanic_data['Age'] >= 18)]

# 3. Sort adult males by descending 'Fare_per_class'
adult_males_srt = adult_males.sort_values(by='Fare_per_class', ascending=False)

# 4. For Select Name, Age, and Fare_per_class columns
result = adult_males_srt[['Name', 'Age', 'Fare_per_class']].head(5)

# 5. Display the result
result
```

# <font color = "Blue">3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

**<font color = "red">What percent of the total fare revenue came from each passenger class?**

```python
# 1. Calculate the total Fare paid across all passengers in the Titanic dataset.
# 2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
# 3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).
# 4. Combine the fare totals from first, second, and third classes into a list.
# 5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.
# Based on the dataset Answer the following question:

# Calculate the total fare paid across all passengers.
total_fare = titanic_data['Fare'].sum()

# Subset for passengers in first class and calculate their total fare.
first_class_fare = titanic_data[titanic_data['Pclass'] == 1]['Fare'].sum()

# Subset for passengers in second class and calculate their total fare.
second_class_fare = titanic_data[titanic_data['Pclass'] == 2]['Fare'].sum()

# Subset for passengers in third class and calculate their total fare.
third_class_fare = titanic_data[titanic_data['Pclass'] == 3]['Fare'].sum()

# Combine the fare totals into a list.
class_fares = [first_class_fare, second_class_fare, third_class_fare]

# Calculate the proportion of fare revenue by class.
fare_proportions = [fare / total_fare for fare in class_fares]

# Print the results (optional).
print("Total Fare:", total_fare)
print("First Class Fare:", first_class_fare)
print("Second Class Fare:", second_class_fare)
print("Third Class Fare:", third_class_fare)
print("Fare Proportions by Class:", fare_proportions)
```

**<font color = "red">What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?**

```python
# 1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age 18{64), and "senior" (age 65 and above).
# 2. Calculate the total number of passengers on the Titanic.
# 3. Count the number of passengers in each age group.
# 4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
# in each age group.
# 5. Display the proportion as a percentage.

# Adding Titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Worksheet-2/Dataset week3/Copy of Titanic-Dataset.csv')

# 1. Create 'age_group' column
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

titanic_data['age_group'] = titanic_data['Age'].apply(categorize_age)

# 2. Calculate total passengers
total_passengers = len(titanic_data)

# 3. Count passengers in each age group
age_group_counts = titanic_data['age_group'].value_counts()

# 4. Calculate proportions
age_group_proportions = age_group_counts / total_passengers

# 5. Display proportions as percentages
print("Age Group Proportions:")
age_group_proportions * 100
```
