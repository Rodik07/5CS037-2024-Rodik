**<Font color = "Red">Task - 3 - Outlier Detection:**

```python
# Calculate IQR for GDP per Capita and Score
Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define outlier boundaries
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outliers
outliers = south_asia_df[
    (south_asia_df['Log GDP per capita'] < lower_bound_gdp) |
    (south_asia_df['Log GDP per capita'] > upper_bound_gdp) |
    (south_asia_df['score'] < lower_bound_score) |
    (south_asia_df['score'] > upper_bound_score)
]

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, hue=south_asia_df.index.isin(outliers.index), palette={True: 'red', False: 'blue'}) #Highlight outliers
plt.title('Outlier Detection in South Asian Countries')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Discuss the characteristics of outliers and their potential impact (add your analysis here)
print("Outlier Countries:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define South Asian countries and load the dataset (if not already loaded)
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
file_path = '/content/drive/MyDrive/Assignment/WHR-2024-5CS037.csv'
df = pd.read_csv(file_path)

# Create south_asia_df by filtering the main DataFrame
south_asia_df = df[df['Country name'].isin(south_asian_countries)]

# Define Middle Eastern countries and create middle_east_df
middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
                         "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
                         "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]
middle_east_df = df[df['Country name'].isin(middle_east_countries)]

# Calculate IQR for South Asia and Middle East
south_asia_q1 = south_asia_df['score'].quantile(0.25)
south_asia_q3 = south_asia_df['score'].quantile(0.75)
south_asia_iqr = south_asia_q3 - south_asia_q1

middle_east_q1 = middle_east_df['score'].quantile(0.25)
middle_east_q3 = middle_east_df['score'].quantile(0.75)
middle_east_iqr = middle_east_q3 - middle_east_q1

# Detecting outliers using IQR
south_asia_outliers = south_asia_df[(south_asia_df['score'] < (south_asia_q1 - 1.5 * south_asia_iqr)) |
                                    (south_asia_df['score'] > (south_asia_q3 + 1.5 * south_asia_iqr))]

middle_east_outliers = middle_east_df[(middle_east_df['score'] < (middle_east_q1 - 1.5 * middle_east_iqr)) |
                                      (middle_east_df['score'] > (middle_east_q3 + 1.5 * middle_east_iqr))]

# Plotting outliers for South Asia and Middle East
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_outliers, color='red')
plt.title('Outliers Based on Score and GDP per Capita (South Asia)')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_outliers, color='blue')
plt.title('Outliers Based on Score and GDP per Capita (Middle East)')
plt.show()
```
